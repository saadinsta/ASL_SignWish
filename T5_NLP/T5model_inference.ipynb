## Importing required libraries

```python
import pandas as pd
from transformers import T5Tokenizer, T5ForConditionalGeneration
import warnings
warnings.filterwarnings('ignore')
# from transformers.optimization import  Adafactor 
import torch
```

```python
if torch.cuda.is_available():
    dev = torch.device("cuda:0") 
    print("Running on the GPU")
else:
    dev = torch.device("cpu")
    print("Running on the CPU")
```

```python
# Load saved model
# tokenizer = T5Tokenizer.from_pretrained('t5-base')

# t5model = T5ForConditionalGeneration.from_pretrained(
#     './pretrained_models/T5_pretrainedmodel_2021-10-26.bin', 
#     return_dict=True,
#     config='./T5modelconfig/t5-base-config.json'
#     )

# Load small model
tokenizer = T5Tokenizer.from_pretrained('t5-small')

t5model = T5ForConditionalGeneration.from_pretrained(
    './pretrained_models/T5small_pretrainedmodel_2021-10-26_v2.bin', 
    return_dict=True,
    config='./T5modelconfig/t5-small-config.json'
    )
```

```python

def generate(list_text,model,tokenizer):
   model.eval()
   
   list_text_new = []
   for word in list_text:
      if ' ' in word:
         word = word.replace(' ', '_')
      list_text_new.append(word)

   text = ' | '.join(list_text_new)

   input_ids = tokenizer.encode("WebNLG:{} </s>".format(text), 
                               return_tensors="pt")  
   input_ids=input_ids.to(dev)
   outputs = model.generate(input_ids)
   result = tokenizer.decode(outputs[0])[:-4].split()[1:]

   return ' '.join(result)
```

```python
l = ['how much', 'laptop', 'cost']
lnew = []
for word in l:
    if ' ' in word:
        word = word.replace(' ', '_')
    lnew.append(word)
' | '.join(lnew)
```

```python
generate(['how_much', 'phone', 'cost'], t5model, tokenizer)
```

```python
generate([''], t5model, tokenizer)
```
